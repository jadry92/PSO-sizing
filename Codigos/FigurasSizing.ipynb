```python
def align_yaxis(ax1, ax2):
    y_lims = np.array([ax.get_ylim() for ax in [ax1, ax2]])

    # force 0 to appear on both axes, comment if don't need
    y_lims[:, 0] = y_lims[:, 0].clip(None, 0)
    y_lims[:, 1] = y_lims[:, 1].clip(0, None)

    # normalize both axes
    y_mags = (y_lims[:,1] - y_lims[:,0]).reshape(len(y_lims),1)
    y_lims_normalized = y_lims / y_mags

    # find combined range
    y_new_lims_normalized = np.array([np.min(y_lims_normalized), np.max(y_lims_normalized)])

    # denormalize combined range to get new axes
    new_lim1, new_lim2 = y_new_lims_normalized * y_mags
    ax1.set_ylim(new_lim1)
    ax2.set_ylim(new_lim2)



nxaxis=['PV=0','0<PV<Demand', 'PV>Demand'];

D=[70,50, 0];
ee=[0,0,0];
le=[58,46, 0];
dlc=[52,42, 0];
dg=[44,38, 0];
pv=[0,10, 0];
b=[14,0, 0];

ee2=[0,0,75];
pv2=[0,0, 68];
D2=[0,0, 48];
le2=[0,0, 0];
dlc2=[0,0, 0];
dg2=[0,0, 0];
b2=[0,0, -20];

width = 0.3

xaxis=np.arange(len(nxaxis))+1
fig, ax1 = plt.subplots(figsize=(5, 4))


ax1.bar(xaxis, D,     width=width, alpha=1, label=u'Demand'             ,color='tab:blue')
ax1.bar(xaxis, le,    width=width, alpha=1, label=u'Lack of energy'     ,color='darkorange')
ax1.bar(xaxis, dlc,   width=width, alpha=1, label=u'Direct load control',color='yellowgreen')
ax1.bar(xaxis, dg,    width=width, alpha=1, label=u'Diesel generator'   ,color='skyblue')
ax1.bar(xaxis, b,     width=width, alpha=1, label=u'Bess'               ,color='goldenrod')
ax1.bar(xaxis, pv,    width=width, alpha=1, label=u'PV'                 ,color='mediumorchid')
ax1.bar(xaxis, ee,    width=width, alpha=1, label=u'Excess of energy'     ,color='forestgreen'  )

ax1.yaxis.grid(True)
ax1.set_ylabel('Power dispatch')
#ax1.set_xlabel(u'Possible escenarios')
ax1.set_xticks(xaxis)
ax1.set_xticklabels(nxaxis)
ax1.legend(loc=7, bbox_to_anchor=(1.02, -0.38),
          ncol=2, fancybox=True, shadow=True)



ax2 = ax1.twinx()
ax2.bar(xaxis, ee2,    width=width, alpha=1, label=u'Excess of energy'   ,color='forestgreen'  )
ax2.bar(xaxis, pv2,    width=width, alpha=1, label=u'PV'                 ,color='mediumorchid')
ax2.bar(xaxis, D2,     width=width, alpha=1, label=u'Demand'             ,color='tab:blue')
ax2.bar(xaxis, dg2,    width=width, alpha=1, label=u'Diesel generator'   ,color='skyblue')
ax2.bar(xaxis, dlc2,   width=width, alpha=1, label=u'Direct load control',color='yellowgreen')
ax2.bar(xaxis, le2,    width=width, alpha=1, label=u'Lack of energy'     ,color='darkorange')
ax2.bar(xaxis, b2,     width=width, alpha=1, label=u'Bess'               ,color='goldenrod')


plt.title('Possible escenarios for the dispatch')

fig.tight_layout()
align_yaxis(ax1, ax2)
fig.savefig('dispatchbar.pdf', bbox_inches='tight')
plt.show()
```
